```python
import numpy as np
from codebase.file_utils import (
    save_obj,
    load_obj,
    make_folder,
    path_backslash
)
from codebase.data import get_data
import altair as alt
import pandas as pd

```

```python
log_dir = 'log/20201101_221415_demo2/'
# particles = load_obj('particles', log_dir)
data = load_obj('data', log_dir)
data
```

```python
def get_df(log_dir, t, name):
    a = pd.DataFrame(
        np.squeeze(
            load_obj(
            'particles'+str(t),
                log_dir
            )['alpha']
        ),
        columns=[i for i in range(6)]
    )
    a['t']= t
    return a.melt(id_vars='t', var_name = 'col')

dfs = []
dfs.append(get_df(log_dir, 1, 'alpha'))
dfs.append(get_df(log_dir, 2, 'alpha'))
dfs.append(get_df(log_dir, 5, 'alpha'))
dfs.append(get_df(log_dir, 20, 'alpha'))
ps = pd.concat(dfs)
```

```python
ps['truevalue'] = ps['col'].map(pd.DataFrame(data['alpha']).to_dict()[0])
```

```python
particle_chart = alt.Chart(ps).mark_area(
    opacity = 0.6,
    interpolate = "step"
).encode(
    alt.Y('count()', stack=None, title=None),
    alt.X('value:Q',
          bin=alt.Bin(maxbins=30),
          title=None
         ),
    alt.Color('t:N')
).properties(width=200, height=100)

rule = alt.Chart(ps).mark_rule(color='red').encode(
    x='truevalue',
    size=alt.value(1)
)

(particle_chart + rule).facet(
    alt.Facet('col'),
    columns=3
)
```
