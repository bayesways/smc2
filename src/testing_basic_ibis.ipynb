```python
import numpy as np
from codebase.file_utils import (
    save_obj,
    load_obj,
    make_folder,
    path_backslash
)
from codebase.data import get_data
import altair as alt
import pandas as pd

```

```python
log_dir = 'log/20201101_221415_demo2/'
# particles = load_obj('particles', log_dir)
data = load_obj('data', log_dir)
data
```

```python
def get_df(log_dir, t, name):
    a = pd.DataFrame(
        np.squeeze(
            load_obj(
            'particles'+str(t),
                log_dir
            )['alpha']
        ),
        columns=['col'+str(i) for i in range(1,7)]
    )
    a['t']= t
    return a.melt(id_vars='t')

dfs = []
dfs.append(get_df(log_dir, 1, 'alpha'))
dfs.append(get_df(log_dir, 10, 'alpha'))
dfs.append(get_df(log_dir, 20, 'alpha'))
ps = pd.concat(dfs)
```

```python
alt.Chart(ps).mark_area(
    opacity = 0.6,
    interpolate = "step"
).encode(
    alt.Y('count()', stack=None),
    alt.X('value:Q',
          bin=alt.Bin(maxbins=30)
         ),
    alt.Color('t:N'),
    column = 'variable'
).properties(width=200, height=100)
```
